```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Load your CSV data into a DataFrame
df = pd.read_csv('Dr.csv')

# Assume 'x', 'y', 'z' are the column headers for your easting, northing, and elevation data
points = df[['x', 'y']].values
values = df['z'].values

# Create a grid where you want to interpolate
grid_x, grid_y = np.mgrid[min(df['x']):max(df['x']):100j, min(df['y']):max(df['y']):100j]

# Interpolate the values
grid_z = griddata(points, values, (grid_x, grid_y), method='cubic')

# Plot the result as a DEM
plt.figure()
cp = plt.contourf(grid_x, grid_y, grid_z, cmap='terrain')
plt.colorbar(cp)
plt.title('Digital Elevation Model')
plt.xlabel('Easting')
plt.ylabel('Northing')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
from mpl_toolkits.mplot3d import Axes3D

# Load your CSV data into a DataFrame
df = pd.read_csv('Dr.csv')

# Assume 'x', 'y', 'z' are the column headers for your easting, northing, and elevation data
points = df[['x', 'y']].values
values = df['z'].values

# Create a grid where you want to interpolate
grid_x, grid_y = np.mgrid[min(df['x']):max(df['x']):100j, min(df['y']):max(df['y']):100j]

# Interpolate the values
grid_z = griddata(points, values, (grid_x, grid_y), method='cubic')

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(grid_x, grid_y, grid_z, cmap='terrain', linewidth=0, antialiased=False)

# Add contour lines
contours = ax.contour(grid_x, grid_y, grid_z, zdir='z', offset=np.min(grid_z), cmap='terrain')

# Add a color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

# Set labels
ax.set_title('3D Digital Elevation Model with Contour Lines')
ax.set_xlabel('Easting')
ax.set_ylabel('Northing')
ax.set_zlabel('Elevation')

# Show plot
plt.show()
```
